```python
from opencage.geocoder import OpenCageGeocode
import pandas as pd
import json
import requests
from config import api_key
```

```python
# The path to our CSV file
file = "../Data Sets/api1.csv"

# Read our Kickstarter data into pandas
investor_df = pd.read_csv(file)
investor_df.head()
```

```python
lat_list = []
lon_list =[]
```

```python
geocoder = OpenCageGeocode(api_key)
```

```python
for index, row in investor_df.iterrows():
    city = row["city"] + ", "
    if row["country"] == "USA":
        state = row["state"] + ", "
    else:
        state = ""
    country = row["country"]
    
#     print(f"{city}{state}{country}")
#     build query url
    query = str(city)+str(state)+str(country)

    try:
        results = geocoder.geocode(query)   
        lat = results[0]["geometry"]["lat"]
        lon = results[0]["geometry"]["lng"]

        lat_list.append(lat)
        lon_list.append(lon)

#         print(lat)
#         print(lon)
#         print("-------")
    except:
        print(f"Coordinates for {query} Not Found")
        print("-------")
        lat_list.append(0)
        lon_list.append(0)
        pass
```

```python
print(len(lat_list))
print(len(lon_list))
```

```python
# create new columns from lists    

investor_df['lat'] = lat_list   
investor_df['lon'] = lon_list

investor_df.head()
```
